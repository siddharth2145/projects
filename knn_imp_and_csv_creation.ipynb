```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("GDP.csv")
```

```python
df.head()
```

```python
selected_columns=df.iloc[:, [0, 1, -1]]#to acces the columns-12,last
```

```python
selected_columns
```

```python
from sklearn.impute import KNNImputer
```

```python
numeric_cols = df.select_dtypes(include='number').columns#to select only numeric values
```

```python
imputer=KNNImputer(n_neighbors=3)#knn imputer
GDPI=imputer.fit_transform(df[numeric_cols])
```

```python
array=pd.DataFrame(GDPI)#to convert it into n-array into a dataframe
```

```python
array
```

```python
combined_df = pd.concat([selected_columns,array], axis=1)#to combine data frames horizontally
```

```python
combined_df
```
